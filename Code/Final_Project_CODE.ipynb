# Lag Determination - BIC Criterion Determination

The Bayesian Information Criterion (BIC) is a statistical criterion used for model selection, including the determination of the lag length in time series models such as Vector Autoregression (VAR). 
It is designed to balance the goodness of fit of a model with its complexity, penalizing overly complex models.

The formula for BIC is given by:

BIC = −2×ln(L)+k×ln(n)

where:

L: the likelihood of the model,

k: the number of estimated parameters in the model,

n: the number of observations.

```python
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import grangercausalitytests
from statsmodels.tools.eval_measures import bic
from itertools import product
```

```python
print(merged.dtypes)
```

```python
merged2 = merged.copy()

# Convert the date column to datetime type
merged2['date'] = pd.to_datetime(merged2['date'])

# Set the date column as the index
merged2.set_index('date', inplace=True)
merged2
```

```python
# I defined the function to calculate BIC:
def bic(log_likelihood, num_params, num_obs):
    return -2 * log_likelihood + num_params * np.log(num_obs)

# I defined the function to perform Granger causality test and calculate BIC:
def granger_causality_bic(data, max_lag):
    best_bic = np.inf
    best_lag = 0
    bic_values = []  # to store BIC values for each lag
    
    for lag in range(1, max_lag + 1):
        # For fitting Granger causality test
        result = grangercausalitytests(data, lag, verbose=False)
        
        # For extracting log-likelihood and number of parameters for BIC calculation
        log_likelihood = result[lag][0]['lrtest'][0]
        num_params = lag * len(data.columns)
        
        # For calculating BIC
        current_bic = bic(log_likelihood, num_params, len(data))
        
        # For updating the best lag if current BIC is better
        if current_bic < best_bic:
            best_bic = current_bic
            best_lag = lag
        
        bic_values.append(current_bic)  # store BIC value for the current lag
    
    return best_lag, best_bic, bic_values

# Determining the optimal lag using BIC for our model:
max_lag_to_test = 10  # I set the max lag to 10 (the results don't change when we change the max lag)
optimal_lag, optimal_bic, bic_values = granger_causality_bic(merged2, max_lag_to_test)

# Print BIC values for each lag
print("BIC values for each lag length:")
for lag, bic_value in enumerate(bic_values, 1):
    print(f"Lag {lag}: {bic_value}")

print("\nOptimal BIC value:", optimal_bic)
print("Optimal lag determined by BIC:", optimal_lag)
```

# CAUSAL RELATIONSHIP

**The Granger Causality test** is used to determine whether or not one time series is useful for forecasting another.

This test uses the following null and alternative hypotheses:

**Null Hypothesis (H0):** "Unemployment time series" does not Granger-cause "inflation time series"

**Alternative Hypothesis (HA):** "Unemployment time series" Granger-causes "inflation time series"

The term “Granger-causes” means that knowing the value of "unemployment time series" at a certain lag is useful for predicting the value of "inflation time series" at a later time period.

This test produces an F test statistic with a corresponding p-value. If the p-value is less than a certain significance level (i.e. α = .05), then we can reject the null hypothesis and conclude that we have sufficient evidence to say that time series x Granger-causes time series y.

# 1) Granger Causality

```python
#Step 1: Load the Data
merged
```

```python
#Step 2: Perform the Granger-Causality Test
from statsmodels.tsa.stattools import grangercausalitytests

grangercausalitytests(merged[['inflation', 'unemployment_rate']], maxlag=[1])
#y: inflation // X: unemployment rate
```

The F test statistic turns out to be 2.7931 and the corresponding p-value is 0.0959.

Since the p-value is greater than .05, we can't reject the null hypothesis of the test and conclude that knowing the unemployment rate is NOT useful for predicting the future inflation rate.

Thus, we can say that "unemployment time series" DOES NOT Granger-causes "inflation time series".

```python
#Step 3: Perform the Granger-Causality Test in Reverse
grangercausalitytests(merged[['unemployment_rate', 'inflation']], maxlag=[1])
#y: unemployment rate // X: inflation
```

The F test statistic turns out to be 5.1889 and the corresponding p-value is 0.0236.

Since the p-value is less than .05, we reject the null hypothesis of the test and conclude that knowing the inflation rate is useful for predicting the future unemployment rate. 

Thus, we can say that "inflation time series" DOES Granger-causes "unemployment time series".

I our data:

2) **independent variable:** inflation

3) **insturmental variable:** national_defense


Finding a valid instrumental variable is not easy. You need to find a variable that satisfies two conditions: **relevance** and **exogeneity.** 

# 1. Relevance Condition

**Relevance** means that the instrumental variable is strongly correlated with the explanatory variable of interest

```python
# Assume that 'dependent', 'independent', and 'instrumental' are the column names
dependent_var = 'unemployment_rate'
independent_var = 'inflation'
instrumental_var = 'national_defense'

# Add a constant term to the independent variable (for the intercept)
merged_all['constant'] = 1

# Fit a simple regression model for the independent and instrumental variable
model = sm.OLS(merged_all[independent_var], merged_all[[instrumental_var, 'constant']])
results = model.fit()

# Check for relevance
relevance_condition = results.f_pvalue < 0.05  # Adjust the significance level as needed
print(results.f_pvalue)

if relevance_condition:
    print("Instrumental variable is relevant.")
else:
    print("Instrumental variable is not relevant.")
```

# 2. Exogeneity

**Exogeneity** means that the instrumental variable is independent of the error term and does not affect the outcome variable, through other channels.

Exogeneity requires that **Cov(instrumental_var, error term)=0.** This cannot be tested.

**Reason:** argument that it is driven by political and strategic factors and not by economic conditions.
